# Create a wordcloud for categories


```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
text = df['categories'].to_string()
```

```python
text = text.lower()
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["restaurants","restaurant","food"])
```

```python
wordcloud = WordCloud(stopwords=stopwords, max_font_size=50, max_words=1000, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
